```python
df["date"]=pd.to_datetime(df["date"])
df["date"]
```

```python
#Total Data analysis for different columns

Total_cases=df["new_cases"].sum()
print("Total_cases:",Total_cases)

Total_deaths=df["new_deaths"].sum()
print("Total_deaths:",Total_deaths)

Total_tests=df["new_tests"].sum()
print("Total_tests",Total_tests)

Total_Vaccinations=df["new_vaccinations"].sum()
print("Total_vaccination", Total_Vaccinations)

Average_life_expectancy=df["life_expectancy"].mean()
print("Average_life_expectancy",Average_life_expectancy)
```

```python
Total_boosters= df["total_boosters"].sum()
print(Total_boosters)
```

```python
Total_populations=df["population"].sum()
Total_populations
```

```python
#Creating Sum New Columns according to needs

#Death Percentage Calculation
df.loc[:, "death_percentage"]= (df["total_deaths"]/df["total_cases"])*100

# COVID Percentage calculation
df.loc[:, "covid_percentage"] = (df["total_cases"] / df["population"]) * 100

# Vaccination Percentage calculation
df.loc[:, "vaccination_percentage"] = (((df["new_vaccinations"]) / 2) / df["population"]) * 100
```
